```python
# 7.1 Weather You Like It Or Not
# Complete the weather application below

API_key = ''


command = ""
while(command != "q"):
    command = input("Choose [c] for current weather, [f] for a forecast, and [q] to quit.")
    if(command == "c"):
        city_name = input("What location would you like the current weather for: ")
        # A. Use the location string to send a search query to this endpoint: 
        # https://openweathermap.org/current#one
        address = f"https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_key}&units=metric"
        data = requests.get(address)
        json_file = data.json()
        temperature = json_file['main']['temp']
        
        print(f"The weather in {city_name} is {temperature} degrees")
        
    if(command == "f"):
        location = input("What location would you like a forecast for?")
        count = int(input("How many timestamps would you like (choose up to 40)"))
        # B. Use the location and count variables for a request to this endpoint: 
        address = f'https://api.openweathermap.org/data/2.5/forecast?q={location}&appid={API_key}&units=metric'

        data = requests.get(address)
        json_file = data.json()
        json_file

        for i in range(0, count):
            temp = json_file['list'][i]['main']['temp']
            weather_description = json_file['list'][i]['weather'][0]['description']
            date = json_file['list'][i]['dt_txt']
            print(f"The weather in {city_name} on {date} will be {weather_description} with {temp} degres Celsius")


    
```

```python
# A. How many Yen will my $100 US Dollars convert to at today's exchange rates?
# https://github.com/fawazahmed0/currency-api#readme

currency_address = 'https://cdn.jsdelivr.net/gh/fawazahmed0/currency-api@1/latest/currencies/usd.json'
data_currency = requests.get(currency_address).json()
exchange_rate = data_currency['usd']['jpy']

usd = 100
jpy = usd * exchange_rate 

print(jpy)
```

```python
# D. Find me a job posting to help my unemployed Graphic Designer friend.
jobs = 'https://arbeitnow.com/api/job-board-api'
jobs_request = requests.get(jobs).json()
jobs_request['data'][1]['title']

for i in range(0,len(jobs_request['data'])):
    if 'graphic designer' in jobs_request['data'][i]['title'].lower():
        print(jobs_request['data'][i]['title'])
        print(jobs_request['data'][i]['description'])
        print(jobs_request['data'][i]['location'])
        print('Remote:', jobs_request['data'][i]['remote'])
    else:
        pass
```
