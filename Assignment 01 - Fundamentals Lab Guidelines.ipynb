```python
import pandas as pd
import random
from itertools import permutations

# Download csv file with the question base
df = pd.read_csv('standardised test.csv')
```

```python
# The data is structured in a csv file with the columns : Question, Right Answer, Answer2, Answer3, Answer4, CAtegory to which the question belongs to.

intent = ""
# Break the program if one answers q
while intent != "q":
    
    intent = input("Welcome! \nWould you like to : \n\n[start] start a new session? \n[add] add new questions to the test? \n\n")

    if intent.lower() == 'start':
        ask_questions()
        
    elif intent.lower() =='add':
        add_question()
        
    else:
        intent = input("Sorry I didn't get it. \nWould you like to : \n[start] start a new session? \n[add] add new questions to the test? \n")

```

```python
# Ask Questions Function 
def ask_questions():
    """
    Will first ask what subject the student wants to train on
    Then proceed to ask each questions until all have been replied to correctly  
    

    """

    # Find the subject the student want to work on
    subject = input("\nWhich subject would you like to train in? \n [Nat] for Nature \n [Geo] for Geography \n [Gas] for Gastronomy\n [His] for History \n [All] for a mix\n")

    if subject.lower() == 'nat':
        questions = df[df['Category'] == 'Nature']
    elif subject.lower() == 'geo':
        questions = df[df['Category'] == 'Geography']
    elif subject.lower() == 'gas':
        questions = df[df['Category'] == 'Gastronomy']  
    elif subject.lower() == 'his':
        questions = df[df['Category'] == 'History']   
    elif subject.lower() == 'all':
        questions = df
    else:
        subject = input("There might be a typo\nWhich subject would you like to train in? \n [Nat] for Nature \n [Geo] for Geography \n [Gas] for Gastronomy\n [All] for a mix\n")

    # Reset index
    questions = questions.reset_index(drop=True)
    
    # Maybe the person wants to train only... and we might end up with 100s of questions in the end. 
    # Let's ask how many questions the person wants to answer
    questions_quantity = input(f"There are {len(questions)} questions in that category, how many do you want to answer?")
    
    # Then, select a random X questions in that category and create the list of questions to ask (remove the random indexes)
    question_index = random.sample(range(0, len(questions)), (len(questions) - int(questions_quantity)))

    for number in question_index:
        questions = questions.drop([number])
    

     # Reset index
    questions = questions.reset_index(drop=True)
    
    # Create an empty submit variable to start the questions
    submit = ""

    for row in questions:

        # Since we are removing the questions that have been answered properly, we will play until all questions have been answered
        while(len(questions)!=0):

            # We want to find the questions randomly by index, create a random index and find the question.
            random_index = random.randint(0, len(questions)-1)
            question = questions['Question'][random_index]


            # The right answer is always on the columns ('Right Answer'). That'd make the test too easy to answer!
            # Create a list with a random order so we don't know which is right

            # 1. Create a list with the answers in current order
            answers = [questions['Right Answer'][random_index], questions['Answer 2'][random_index], questions['Answer 3'][random_index], questions['Answer 4'][random_index]]

            # Permutate the answers available and choose one option randomly based on the lenght of the permutations
            answers_permutated_list = list(permutations(answers))
            random_answer_index = random.randint(0, len(answers_permutated_list)-1)

            # Create the variable answers that will display the answers on the defined random order
            answers = answers_permutated_list[random_answer_index]
            submit = input(f"\n{question}\n A - {answers[0]}\n B - {answers[1]}\n C - {answers[2]}\n D - {answers[3]}\n ")

            # Check if the answer is right based on the answer given
            if submit.lower() == 'a':
                index = 0
            if submit.lower() == 'b':
                index = 1
            if submit.lower() == 'c':
                index = 2
            if submit.lower() == 'd':
                index = 3

            answer_given = answers[index]


            if answer_given == questions['Right Answer'][random_index]:
                print("\n Well Done!")
                questions = questions.drop(random_index)
                questions = questions.reset_index(drop=True)
            else:
                print('\n Wrong, try again!')

        print('\n All questions have been answered properly, nice one!')
        break
```

```python
# For the ones that want to add a question...

def add_question():
    """
    Will ask the student to enter each questions, answers and categories
    Will store this informations onto the already existing dataframe
    
    """
    
    
    if intent.lower() == 'add':

        new_question = input("Enter your question here: ")
        right_answer = input("Enter the right answer here: ")
        answer2 = input("Enter a random answer here: ")
        answer3 = input("Enter a random answer here: ")
        answer4 = input("Enter a random answer here: ")
        category = input("What is the category? ")

        df.loc[len(df.index)] = [new_question, right_answer, answer2, answer3, answer4, category]

        return df
```
