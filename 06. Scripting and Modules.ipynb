```python
# C. Simulate a slot machine with the following rules:
#    i. Each turn, the player loses 1 point and three random elements are chosen
#    ii. If the three elements are all the same, the player wins 2 points
#    iii. If all three elements are different, the player wins 5 points --> changed to loose
#    Continue taking turns until the player has 0 points or more than 10 points
#    Each turn, print a message describing the outcome of the machine

elements = ["red", "green", "blue"]

player_points = 7


# Find the lenght of the string to set up the randint range
elements_lenght = len(elements)

# Set up the random results for each of the three elements separately


while 0 < player_points < 10:
    elements1 = random.randint(0, elements_lenght-1)
    elements2 = random.randint(0, elements_lenght-1)
    elements3 = random.randint(0, elements_lenght-1)
    print(f"And this round's result is... {elements1}-{elements2}-{elements3}!")
    if elements1 == elements2 and elements2 == elements3:
        player_points += 2
        print('You win, play again!')
    else:
        player_points -= 5
        print('You loose!')
        
if player_points<0:
    print('There is no more money in your account!')
else:
    print('All the millions to you')
    
```

```python
# 6.2 GeoCoder Road Trip
destinations = ["Space Needle",
                "Crater Lake",
                "Golden Gate Bridge",
                "Yosemite National Park",
                "Las Vegas, Nevada",
                "Grand Canyon National Park",
                "Aspen, Colorado",
                "Mount Rushmore",
                "Yellowstone National Park",
                "Sandpoint, Idaho",
                "Banff National Park",
                "Capilano Suspension Bridge"]

# A. Install and import the geocoder library: https://geocoder.readthedocs.io/index.html
#    Then, print a statement describing the latitude and longitude of our first destination, Space Needle

place = geocoder.location(destinations[0])
print(f" The {destinations[0]} is located at {place}")
```

```python
# B. Find the farthest south destination (the result with the lowest latitude)

#set up an initial latitude to the max that exists +1 : North Pole
latitude = 91
location = "North Pole"

for destination in destinations:
    
    #Find the geocoordinates of that area and set up the latitude
    place = geocoder.location(destination)
    new_latitude = place.lat
    
    # If the new_latitude is the lowest, set it up as the new standard "latitude" and location
    if new_latitude < latitude:
        latitude = new_latitude
        location = destination
        
    # Otherwise, just continue on that loop
    else:
        pass

print(f"The farthest South Location is {location} at latitude {latitude}")
```

```python
# C. For each destination, describe the direction being traveled from the previous destination
#    Hint: You can keep track of where we are using current_lat and current_lng variables
#    If the next destination has a greater latitude, it means we're going North
#    If the next destination has a greater longitude, it means we're going West
#    Print a message in the following format: "Next, we travel SouthWest to XXXXX"

# Setup Starting Point
starting_point = "Chicago"
starting_latitude = geocoder.location(starting_point).lat
starting_longitude = geocoder.location(starting_point).lng

print(f"Leaving for {starting_point}, let's go travel!")

for destination in destinations:

    #Find the geocoordinates of that area and set up the latitude
    place = geocoder.location(destination)
    destination_latitude = place.lat
    destination_longitude = place.lng
    
    # Logic to for the North/South and East/West directions:
    if destination_latitude > starting_latitude:
        direction_NS = "North"
    else:
        direction_NS = "South"
    if destination_longitude > starting_longitude:
        direction_EW = "West"
    else:
        direction_EW = "East"
    
    print(f"Next, we travel {direction_NS}-{direction_EW} to {destination}")
    
    # Reset the starting point for the next travel
    starting_latitude = destination_latitude 
    starting_longitude = destination_longitude 

```
